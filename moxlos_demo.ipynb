```python
import ipywidgets as widgets

from sympy import Symbol, Derivative, sympify, pprint, init_printing, lambdify
from sympy.core.sympify import SympifyError

from numpy import arange
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = open("logo.png", "rb")
image = file.read()
im = widgets.Image(
    value=image,
    format='png',
    width=300,
    height=400,
)


fun = widgets.Text(
    value='x**2',
    placeholder='Type something',
    description='Function:',
    disabled=False
)



var = widgets.Text(
    value='x',
    placeholder='Type something',
    description='Variable:',
    disabled=False
)



a = widgets.BoundedFloatText(
    value=0.0,
    min=-10.0,
    max=10.0,
    step=0.1,
    description='Lower Bound:',
    disabled=False
)


b = widgets.BoundedFloatText(
    value=1.0,
    min=-10.0,
    max=10.0,
    step=0.1,
    description='Upper Bound:',
    disabled=False
)


w = widgets.IntSlider(
    value=100,
    min=10,
    max=500,
    step=10,
    description='Step:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)





def derivative(fun, var):
    
    init_printing()

    d = Derivative(fun, var).doit()
    pprint(d)

def plot_funct(f,var,a,b,mesh_size):
    
    x = arange(a,b,(b-a)/mesh_size)

    var = Symbol(var)
    d = Derivative(f, var).doit()
    f_sym = Derivative(f,var,0).doit()
    y_f = lambdify(var, f_sym)
    y_der = lambdify(var, d)
    y_f = [y_f(x_var) for x_var in x]
    y_der = [y_der(x_var) for x_var in x]



    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,5), dpi=100)
    fig.suptitle('Plots')
    ax1.plot(x, y_f)
    ax2.plot(x, y_der)
    ax1.set_title('function')
    ax2.set_title('derivative')
    plt.show()
    

#Create your button.
button = widgets.Button(description="Derivative",
                       icon='fa-cube')
button_2 = widgets.Button(description="Plot",
                          icon='fa-area-chart')
#Output field.
output = widgets.Output()

#function to handle input.
def showOutput(btn):
    output.clear_output()
    with output:
        derivative(fun.value, var.value)

def showOutput_2(btn):
    output.clear_output()
    with output:
        plot_funct(fun.value,var.value, a.value,b.value,w.value)
  
        
        
        
button.on_click(showOutput)
button_2.on_click(showOutput_2)



ui = widgets.VBox([im,
                   fun,
                   var,
                   a,b,w,
                   button,
                   button_2,
                   output])
ui

```
